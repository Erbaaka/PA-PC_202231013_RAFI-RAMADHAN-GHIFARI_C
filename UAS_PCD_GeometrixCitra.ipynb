```python
#IMPORT LIBRARY
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#LOAD CITRA
image = cv2.imread('Rafi-Olah.jpg')
```

```python
#DISPLAY CITRA
def display_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i in range(len(images)):
        plt.subplot(2, 3, i+1)
        plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
original = image.copy()
```

```python
#ROTASI CITRA
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
img_rotated = cv2.warpAffine(image, M, (w, h))
```

```python
#RESIZE CITRA
new_height = int(h * 1.5)
new_width = w 
img_resized = cv2.resize(image, (new_width, new_height))
```

```python
#CROP CITRA
img_cropped = image[50:200, 50:200]
```

```python
#FLIP CITRA
img_flipped = cv2.flip(image, 1)
```

```python
#TRANSLASI CITRA
M = np.float32([[1, 0, 50], [0, 1, 100]])
img_translated = cv2.warpAffine(image, M, (w, h))
```

```python
#HASIL OLAH CITRA
images = [original, img_rotated, img_resized, img_cropped, img_flipped, img_translated]
titles = ['Citra Asli', 'Rotasi', 'Resize', 'Cropped', 'Flipped', 'Translasi']
display_images(images, titles)
```
